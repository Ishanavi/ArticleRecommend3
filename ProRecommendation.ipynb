```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("shared_articles.csv")
df2 = pd.read_csv("users_interactions.csv")

df1.head()
```

```python
df2.head(60)
```

```python
df1.columns
```

```python
df1 = df1[df1['eventType']=='CONTENT SHARED']
df1.head()
```

```python
def total_events(df1_rows):
  total_likes = df2[(df2["contentId"] == df1_rows["contentId"]) & (df2['eventType'] == "LIKE")].shape[0]
  total_views = df2[(df2["contentId"] == df1_rows["contentId"]) & (df2['eventType'] == "VIEW")].shape[0]
  total_bookmarks = df2[(df2["contentId"] == df1_rows["contentId"]) & (df2['eventType'] == "BOOKMARK")].shape[0]
  total_follow = df2[(df2["contentId"] == df1_rows["contentId"]) & (df2['eventType'] == "FOLLOW")].shape[0]
  total_comments = df2[(df2["contentId"] == df1_rows["contentId"]) & (df2['eventType'] == "COMMENT CREATED")].shape[0]
  return total_likes+total_views+total_bookmarks+total_follow+total_comments

df1["total_events"] = df1.apply(total_events,axis=1)
df1.head()
```

```python
df1 = df1.sort_values(["total_events"],ascending=[False])
df1.head()
```

```python
def convert_to_lowercase(x):
  if isinstance(x,str):
    return x.lower()
  else:
    return ""

df1["title"] = df1["title"].apply(convert_to_lowercase)
df1.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df1['title'])

from sklearn.metrics.pairwise import cosine_similarity

cosine_sim_2 = cosine_similarity(count_matrix, count_matrix)

df1 = df1.reset_index()
indices = pd.Series(df1.index, index=df1['contentId'])
```

```python
def get_recommendations(contentId, cosine_sim):
    idx = indices[contentId]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    article_indices = [i[0] for i in sim_scores]
    return df1['contentId'].iloc[article_indices]
```

```python
get_recommendations(-6783772548752091658, cosine_sim_2)
```
